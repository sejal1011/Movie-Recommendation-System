## Extracting features of 2021 movies from Wikipedia

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2021"
df1 = pd.read_html(link, header=0)[2]
df2 = pd.read_html(link, header=0)[3]
df3 = pd.read_html(link, header=0)[4]
df4 = pd.read_html(link, header=0)[5]
```

```python
import pandas as pd

df_list = [df1, df2, df3, df4]
df = pd.concat(df_list, ignore_index=True)
```

```python
df
```

```python
!pip install tmdbv3api
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = '3c8a48afaa665f0097beed6b27c4191f'
```

```python
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)

    # Check if there are any results
    if result['results']:
        movie_id = result['results'][0]['id']
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id, tmdb.api_key))
        data_json = response.json()

        if data_json['genres']:
            genre_str = " " 
            for i in range(0, len(data_json['genres'])):
                genres.append(data_json['genres'][i]['name'])
            return genre_str.join(genres)
    
    # Return NaN if there are no results or no genres
    return np.NaN
```

```python
df['genres'] = df['Title'].map(lambda x: get_genre(str(x)))
```

```python
df
```

```python
df.columns
```

```python
df_2021 = df[['Title','Cast and crew','genres']]
```

```python
df_2021
```

```python
def get_director(x):
    if pd.isna(x):
        return np.NaN  # Return NaN for NaN values
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]

# Apply the function to the 'Cast and crew' column
df_2021['director_name'] = df_2021['Cast and crew'].map(lambda x: get_director(x))
```

```python
def get_actor1(x):
    if pd.isna(x):
        return np.NaN
    return ((x.split("screenplay); ")[-1]).split(", ")[0])

# Apply the function to the 'Cast and crew' column
df_2021['actor_1_name'] = df_2021['Cast and crew'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    if pd.isna(x) or len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])

# Apply the function to the 'Cast and crew' column
df_2021['actor_2_name'] = df_2021['Cast and crew'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    if pd.isna(x) or len((str(x).split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((str(x).split("screenplay); ")[-1]).split(", ")[2])

# Apply the function to the 'Cast and crew' column
df_2021['actor_3_name'] = df_2021['Cast and crew'].map(lambda x: get_actor3(x))
```

```python
df_2021
```

```python
df_2021 = df_2021.rename(columns={'Title':'movie_title'})
```

```python
new_df21 = df_2021.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
new_df21
```

```python
new_df21['actor_2_name'] = new_df21['actor_2_name'].replace(np.nan, 'unknown')
new_df21['actor_3_name'] = new_df21['actor_3_name'].replace(np.nan, 'unknown')
```

```python
new_df21['movie_title'] = new_df21['movie_title'].str.lower()
```

```python
new_df21['comb'] = new_df21['actor_1_name'] + ' ' + new_df21['actor_2_name'] + ' '+ new_df21['actor_3_name'] + ' '+ new_df21['director_name'] +' ' + new_df21['genres']
```
